# Transacciones
- Desde que se conecta hasta que termina la peticion
- SELECT * FROM TABLA -> conectar consultar, va obtener y te va devolver y ahi muere.
- UPDATE valor -> conecta, va leer el valor, va actualizarlo, va devolver el valor y ahi muere.

## Caso Transacciones 
precio = 10bs,
- si ambos vienen al mismo tiempo.
- si es mayor, igual a 10bs sumale 5bs
- si es mayor, igual de 15bs sumale 5bs

## relaciones en base  de datos

- 1 a 1: para añadir mas informacion y dividir una tabla 
- 1 a n: el 1 es el padre y sus hijos serian los N
- n a n: Ej. autor y libros, en los ORM se tiene que crear un tabla intermedia para registrar las ocurrencias con sus PK 
libro - autor
el principito - carlos
el principito - pablo
- n a [1 1] a n

## estados
leido -> modificacion -> respuesta 
- en caso de exito: commit
- en caso de error: rollback
