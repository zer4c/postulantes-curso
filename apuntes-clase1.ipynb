# producto
- CRUD 
    - create - crear un producto - post
    - read - leer un producto - get
    - read - leer varios productos - get
    - update - actualizar un producto completamente - put 
    - update - actualizar un producto parcialmente - patch
    - delete - eliminar un producto - delete
